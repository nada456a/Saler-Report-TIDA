# Recomendation time display advertising

```python
sale['hour']= sale['Waktu Pembayaran Dilakukan'].str.split(' ').str.get(-1)
sale['hour']=sale['hour'].str.split(':').str.get(0)
sale_adv=sale.groupby('hour')['Nama Produk'].count(). sort_index(ascending=True)
sale_adv.to_frame()
```

have any number duplicate or different format like 04=4, so replace it

```python
sale['hour'].replace('4', '04',  inplace=True)
sale['hour'].replace('6', '06',  inplace=True)
sale['hour'].replace('7', '07',  inplace=True)
sale['hour'].replace('8', '08',  inplace=True)
sale['hour'].replace('9', '09',  inplace=True)
```

```python
mp.figure(figsize=(10,6))
sale_adv=sale.groupby('hour')['Nama Produk'].count(). sort_index(ascending=True)
x= sale_adv.index
y=sale_adv.values
mp.xlabel('Hour')
mp.ylabel('Number of Orders')
mp.grid()
mp.plot(x, y)
```
